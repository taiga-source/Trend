```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
import japanize_matplotlib

# 既存のコード
def splitList_5(listA):
    chunks = [listA[i:i + 5] for i in range(0, len(listA), 5)]
    return chunks

listA = [i for i in range(100)]

casual_styles = [
    "ベーシック系",
    "アメリカジコーデ",
    "モードカジュアル",
    "スポーツカジュアル",
    "ストリート系",
    "アウトドアスタイル"
]

elegant_styles = [
    "エレガンスファッション",
    "ガーリー",
    "オフィスカジュアル",
    "キャリアファッション",
    "マニッシュ",
    "デザイナーズファッション"
]

other_styles = [
    "トラッド",
    "ノームコア",
    "ナチュラル系",
    "韓国系",
    "量産型",
    "地雷系"
]

sum_list = casual_styles + elegant_styles + other_styles

pytrends = TrendReq(hl='ja-JP', tz=360)

list_5 = splitList_5(sum_list)

def creatdf(listA):
    pytrends.build_payload(listA, timeframe='2005-01-01 2024-8-18', geo='JP')
    df = pytrends.interest_over_time()
    return df

list_df = []
for i in range(len(list_5)):
    df = creatdf(list_5[i])
    list_df.append(df)

# すべてのデータフレームを結合
combined_df = pd.concat(list_df, axis=1)

# 重複したカラムを削除
combined_df = combined_df.loc[:,~combined_df.columns.duplicated()]

# グラフの表示
plt.figure(figsize=(15, 10))
sns.lineplot(data=combined_df)
plt.title('ファッショントレンド')
plt.xlabel('日付')
plt.ylabel('関心度')
plt.legend(title='ファッションスタイル', loc='upper left', bbox_to_anchor=(1, 1))
plt.grid(True)
plt.tight_layout()
plt.show()
```
