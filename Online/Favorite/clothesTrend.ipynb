```python
from sklearn.metrics import mean_absolute_error, mean_squared_error

def creat_df(y):
  df2 = pd.DataFrame()
  df2["ds"] = merged_data["date"]
  df2["y"] = merged_data[y]

  df2['merchandise'] = merged_data['merchandise']
  df2['stars'] = merged_data['stars']
  df2['negativeVSpositive'] = merged_data['negativeVSpositive']
  df2['ä¸€ä½'] = merged_data['negativeVSpositive']

  # Prophetãƒ¢ãƒ‡ãƒ«ã®åˆæœŸåŒ–
  model = Prophet()

  # 3ã¤ã®èª¬æ˜å¤‰æ•°ã‚’è¿½åŠ 
  model.add_regressor('merchandise')
  model.add_regressor('stars')
  model.add_regressor('negativeVSpositive')
  model.add_regressor('ä¸€ä½')

  # ãƒ¢ãƒ‡ãƒ«ã®å­¦ç¿’
  model.fit(df2[['ds', 'y', 'merchandise', 'stars', 'negativeVSpositive', 'ä¸€ä½']])

  # äºˆæ¸¬ã™ã‚‹æœŸé–“ã®ãƒ‡ãƒ¼ã‚¿ã‚’ä½œæˆ
  future = model.make_future_dataframe(periods=30)

  # äºˆæ¸¬ã™ã‚‹æœŸé–“ã®èª¬æ˜å¤‰æ•°ã®å€¤ã‚’ä»®è¨­å®šï¼ˆéå»ã®å€¤ã®å¹³å‡å€¤ãªã©ã‚’ä½¿ã†ã¨ã‚ˆã„ï¼‰
  future['merchandise'] = df2['merchandise'].mean()  # ä»®ã®å€¤
  future['stars'] = df2['stars'].mean()  # ä»®ã®å€¤
  future['negativeVSpositive'] = df2['negativeVSpositive'].mean()  # ä»®ã®å€¤
  future['ä¸€ä½'] = df2['ä¸€ä½'].mean()  # ä»®ã®å€¤

  # äºˆæ¸¬ã®å®Ÿè¡Œ
  forecast = model.predict(future)

  # Prophetãƒ¢ãƒ‡ãƒ«ã®åˆæœŸåŒ–
  model = Prophet()

  # èª¬æ˜å¤‰æ•°ã®è¿½åŠ 
  model.add_regressor('merchandise')
  model.add_regressor('stars')
  model.add_regressor('negativeVSpositive')
  model.add_regressor('ä¸€ä½')

  # ãƒ¢ãƒ‡ãƒ«ã®å­¦ç¿’
  model.fit(df2[['ds', 'y', 'merchandise', 'stars', 'negativeVSpositive', 'ä¸€ä½']])

  # å°†æ¥ã®ãƒ‡ãƒ¼ã‚¿ãƒ•ãƒ¬ãƒ¼ãƒ ã‚’ä½œæˆï¼ˆ365æ—¥Ã—6å¹´=6å¹´å…ˆã¾ã§äºˆæ¸¬ï¼‰
  future = model.make_future_dataframe(periods=365*6)

  # å°†æ¥ã®èª¬æ˜å¤‰æ•°ãƒ‡ãƒ¼ã‚¿ã‚’è¿½åŠ ï¼ˆä»®ã«éå»ãƒ‡ãƒ¼ã‚¿ã®å¹³å‡å€¤ã‚’åˆ©ç”¨ï¼‰
  future['merchandise'] = df2['merchandise'].mean()  # ä»®ã®å€¤
  future['stars'] = df2['stars'].mean()  # ä»®ã®å€¤
  future['negativeVSpositive'] = df2['negativeVSpositive'].mean()  # ä»®ã®å€¤
  future['ä¸€ä½'] = df2['ä¸€ä½'].mean()  # ä»®ã®å€¤

  # äºˆæ¸¬
  forecast = model.predict(future)

  # äºˆæ¸¬çµæœã®ãƒ—ãƒ­ãƒƒãƒˆ
  fig1 = model.plot(forecast)

  # ãƒˆãƒ¬ãƒ³ãƒ‰ã¨å­£ç¯€æ€§ã®ãƒ—ãƒ­ãƒƒãƒˆ
  fig2 = model.plot_components(forecast)

  # Prophetãƒ¢ãƒ‡ãƒ«ã®åˆæœŸåŒ–ã¨èª¬æ˜å¤‰æ•°ã®è¿½åŠ 
  model = Prophet()
  model.add_regressor('merchandise')
  model.add_regressor('stars')
  model.add_regressor('negativeVSpositive')
  model.add_regressor('ä¸€ä½')

  # ãƒ¢ãƒ‡ãƒ«ã®å­¦ç¿’
  model.fit(df2[['ds', 'y', 'merchandise', 'stars', 'negativeVSpositive', 'ä¸€ä½']])

  # å°†æ¥ã®ãƒ‡ãƒ¼ã‚¿ãƒ•ãƒ¬ãƒ¼ãƒ ã‚’ä½œæˆï¼ˆ365æ—¥å…ˆã¾ã§äºˆæ¸¬ï¼‰
  future = model.make_future_dataframe(periods=365*6, freq='MS')

  # äºˆæ¸¬æœŸé–“ã«ãŠã‘ã‚‹èª¬æ˜å¤‰æ•°ã®å€¤ã‚’ä»®è¨­å®š
  # éå»ã®å¹³å‡å€¤ã‚„ä¸­å¤®å€¤ã‚’åˆ©ç”¨ã—ã¦ä»®å€¤ã‚’è¨­å®šã§ãã¾ã™
  future['merchandise'] = df2['merchandise'].mean()
  future['stars'] = df2['stars'].mean()
  future['negativeVSpositive'] = df2['negativeVSpositive'].mean()
  future['ä¸€ä½'] = df2['ä¸€ä½'].mean()

  # äºˆæ¸¬ã®å®Ÿè¡Œ
  forecast = model.predict(future)

  # äºˆæ¸¬ç²¾åº¦ã®ç¢ºèª
  # å®Ÿéš›ã®ãƒ‡ãƒ¼ã‚¿ã¨äºˆæ¸¬ãƒ‡ãƒ¼ã‚¿ã®æ¯”è¼ƒ
  predicted = forecast.set_index('ds').loc[df2['ds'], 'yhat']
  actual = df2.set_index('ds')['y']

  mae = mean_absolute_error(actual, predicted)
  mse = mean_squared_error(actual, predicted)

  print(f"Mean Absolute Error (MAE): {mae}")
  print(f"Mean Squared Error (MSE): {mse}")

  # äºˆæ¸¬çµæœã®å¯è¦–åŒ–
  model.plot(forecast)
  model.plot_components(forecast)

  # äºˆæ¸¬ã®å®Ÿè¡Œ
  forecast = model.predict(future)

  # äºˆæ¸¬çµæœã®å¯è¦–åŒ–
  model.plot(forecast)
  model.plot_components(forecast)

  return forecast['yhat'], forecast['ds']

list_modelNeme = ['VAE_MinesystemBasic_z',
                  'VAE_ModeMass_z',
                  'VAE_SportKorea_z',
                  'VEA_StreetNatural_z',
                  'VAE_OutdoorNormcore_z',
                  'VAE_EleganceTrad_z',
                  'VAE_GirlyDesigner_z',
                  'VAE_OfficeManush_z']

df_yhat = pd.DataFrame()
for i in range(len(list_modelNeme)):
  yhat, ds = creat_df(list_modelNeme[i])
  df_yhat['ds'] = ds
  df_yhat[list_modelNeme[i]+'_yhat'] = yhat
```
